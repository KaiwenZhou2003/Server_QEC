```python
import numpy as np
```

```python
mat =np.loadtxt('../results/mat.csv',delimiter=',',dtype=int)
mat
```

```python
nonezeronum = set(np.sum(mat, axis=0).tolist())
mats_cols = {k:[] for k in nonezeronum}
for colidx in range(len(mat[0])):
    mat_col = mat[:, colidx]
    mats_cols[np.count_nonzero(mat_col)].append(mat_col)
np.array(mats_cols[3]).T.shape
```

```python

for k in mats_cols:
    mats_cols[k] = np.array(mats_cols[k]).T
new_mat = np.hstack(list(mats_cols.values()))
print(new_mat)
```
